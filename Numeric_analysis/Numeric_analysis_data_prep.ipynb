# Load Data

Loading numeric data for the year of 2022

```python
nd = pd.read_csv("../datasets/TSLA.csv")
nd
```

```python
nd = nd.loc[:, ['Date','Open','High','Low','Close','Volume']]
nd
```

```python
nd.columns = nd.columns.str.lower()
nd
```

```python
nd['polarity'] = ''
nd
```

```python
for i in range(len(nd) - 1):
    if nd.loc[i+1, 'close'] > nd.loc[i, 'close']:
        nd.loc[i+1, 'polarity'] = 'positive'
    elif nd.loc[i+1, 'close'] < nd.loc[i, 'close']:
        nd.loc[i+1, 'polarity'] = 'negative'
    else:
        nd.loc[i+1, 'polarity'] = 'equal'
```

```python
nd
```

```python
nd.to_csv('../datasets/numeric_data.csv', index=False)
```
