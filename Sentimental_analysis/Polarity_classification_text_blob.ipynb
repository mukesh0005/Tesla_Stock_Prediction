## Load data

```python
import numpy as np
import pandas as pd
```

```python
cs = pd.read_csv("../datasets/sentimental_data.csv")
cs
```

```python
import textblob
from textblob import TextBlob

tweet = cs['tweet']
sentiment = []
for text in tweet:
    analysis = TextBlob(text)
    if analysis.polarity > 0:
        classification = 'positive'
    elif analysis.polarity < 0:
        classification = 'negative'
    else:
        classification = 'neutral'
    sentiment.append(classification)
```

```python
## converting the list to a dataframe and renaming the frame name.
sentiment =  pd.DataFrame(sentiment)
sentiment.rename(columns={0: 'sentiment'}, inplace=True)
sentiment
```

```python
cs_ = pd.concat([cs,sentiment],axis=1)
```

```python
cs_
```

```python
cs_.to_csv('../datasets/Labelled_data.csv', index=False)
```
