## Filtering Columns

Since we only need date, tweet columns selected only those two columns, for extra analysis i have also taken likes so that to determine how likes count can also effect the stock price.

```python
cs = cs.loc[:, ['date','tweet','likes_count']]
cs
```

```python
cs['tweet'] = [item.replace("''",'') for item in cs['tweet']]
```

Removing the mention names of users for tweet replies since this is not useful for the sentiment 

```python
import re
```

```python
tweet = []
for i in range(len(cs)):
    _ = re.sub(r'@[\w]+', '', cs.loc[i,'tweet'])
    _ = re.sub(r'https?://[^\s\n\r]+', '', _)
    _ = re.sub(r'#', '', _)
    tweet.append(_)

cs['tweet'] = tweet
```

```python
cs
```

## Stop word removal

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

for w in cs['tweet']:
	if w not in stop_words:
		filtered_sentence.append(w)
```

```python
filtered_sentence
```

## Word Stemming

Lemmatizing each tweet so that the tweet is reduced to root form.

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag, word_tokenize

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

transformed_corpus = []
wnl = WordNetLemmatizer()
for document in filtered_sentence:
    if isinstance(document, str):
        transformed_document = ""
        for word, tag in pos_tag(word_tokenize(document)):
            wntag = tag[0].lower()
            wntag = wntag if wntag in ['a', 'r', 'n', 'v'] else None
            if not wntag:
                lemma = word
            else:
                lemma = wnl.lemmatize(word, wntag)
            transformed_document += lemma + " "
        transformed_corpus.append(transformed_document)

transformed_corpus
```

## Tokenization and TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Like CountVectorizer, TfidfVectorizer will covert to lowercase, remove punctuation, and remove 
# stop words - to remove other things, such as numbers, use the token_pattern parameter
vectorizer = TfidfVectorizer(stop_words='english', lowercase=True, token_pattern="[^\W\d_]+")

X = vectorizer.fit_transform(transformed_corpus)

df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
df
```

```python
cs_ = pd.concat([cs,df],axis=1)
```
